### Question 4: Data Imputation (3 Marks)
Fill missing numerical values with the median and missing categorical values with the mode. Print the total number of missing values remaining in the entire dataset.



````python
#from sklearn importing Simple Imputer class
from sklearn.impute import SimpleImputer
imputer_median = SimpleImputer(strategy="median")   #for numerical
imputer_mode = SimpleImputer(strategy="most_frequent")  #for categorical

#Selecting numerical and categorical columns seperately
df_num = df.select_dtypes(include = "number")
df_cat = df.select_dtypes(exclude= "number")

#using .fit_transform
num_impute = imputer_median.fit_transform(df_num)
cat_impute = imputer_mode.fit_transform(df_cat)

#Converting to dataframe
df_num_impute = pd.DataFrame(data = num_impute, columns = df_num.columns)
df_cat_impute =  pd.DataFrame(data = cat_impute, columns = df_cat.columns)

#Concatenating
df_impute = pd.concat([df_num_impute,df_cat_impute],axis=1)

df = df_impute.copy()
#checking for null values
df.isnull().sum()
````

### Question 5: Label Encoding (3 Marks)
Use `LabelEncoder` from `sklearn.preprocessing` to encode all categorical columns into numerical values. Print the first 5 rows of the updated DataFrame.

**Hint:** Loop through the categorical columns and apply `.fit_transform()`.

**Expected Output:**
```
A pandas DataFrame head with all columns containing numerical values.
```

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#Dataframe for categorical columns
cat = df.select_dtypes(exclude="number").columns

#Loop through the categorical columns
for i in cat:
  df[i] = le.fit_transform(df[i])
df.head()
````

### Question 6: Feature and Target Separation (3 Marks)
Separate the features `X` and the target `y` (assuming 'performance_category' is the target). Print the shapes of `X` and `y`.

**Expected Output:**
```
Shapes of X and y (e.g., (n_samples, n_features) and (n_samples,)).
```

````python
X = df.drop('performance_category',axis=1)
Y = df['performance_category']
X.shape , Y.shape
````

### Question 7: Train-Test Split (3 Marks)
Split the data into training and testing sets with a test size of 20% and `random_state=42`. Print the shapes of `X_train` and `X_test`.

**Expected Output:**
```
Shapes of X_train and X_test.
```

````python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape, X_test.shape
````

### Question 8: Feature Scaling (3 Marks)
Scale the features using `MinMaxScaler` from `sklearn.preprocessing`. Fit on `X_train` and transform both `X_train` and `X_test`. Print the minimum and maximum values of the scaled `X_train`.



````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#using .fit_transform
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#printing min and max value
print(f'Minimum value of scaled X_train is {X_train_scaled.min()}')
print(f'Maximum value of scaled X_train is {X_train_scaled.max()}')
````

### Question 9: Decision Tree Classifier (5 Marks)
Initialize and train a `DecisionTreeClassifier` with `random_state=42`. Make predictions on the test set and print the accuracy score.

**Expected Output:**
```
A numerical value representing the accuracy score.
```

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier(random_state = 42)

#Decision Tree fitting and predicting
dt.fit(X_train_scaled,Y_train)
y_pred = dt.predict(X_test_scaled)

#printing the Accuracy score
a = accuracy_score(Y_test,y_pred)
print(f'Accuracy Score is {a}')
````

### Question 10: Feature Importance (4 Marks)
Extract the feature importances from the trained Decision Tree model. Print a dictionary or a pandas Series mapping feature names to their importance scores.

**Hint:** Access `.feature_importances_` on the trained model and combine it with `X.columns`.

**Expected Output:**

A mapping of feature names to their importance scores.

````python
feature_importances = pd.Series(data=dt.feature_importances_, index=X_train.columns).sort_values(ascending=False)

#Pandas Series
print("Feature Importances:")
print(feature_importances)
````
